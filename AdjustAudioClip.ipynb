```
import os
import wave
from tqdm import tqdm
from wave import Wave_read
from wave import Wave_write
import numpy as np
from google.colab import drive
dataPath = '/content/drive/My Drive/Dataset'
drive.mount('/content/drive/')
```

```
inputRoute = dataPath+'/AudioClips/FSDKaggle2019.audio_test/'
s = os.listdir(inputRoute)
theVal = len(s)
theVal
totalSamples = 44100
```

```
outputRoute = dataPath+'/OneSecondAudioClips/Audio_Test/'
```

```
os.path.isfile('/content/drive/My Drive/Dataset/AudioClips/FSDKaggle2019.audio_test/382cd8c5.wav')
```

```
index = 0
pbar = tqdm(total=theVal)
while index < theVal:
  fn = s[index]
  theAudio = inputRoute+fn
  f = wave.open(theAudio,'rb')
  p = f.readframes(nframes=totalSamples)
  f.close()
  d = outputRoute+fn
  l = wave.open(d, 'wb')
  l.setnchannels(1)
  l.setsampwidth(4)
  l.setframerate(totalSamples)
  l.writeframesraw(p)
  l.close()
  index += 1
  pbar.update(1)  
pbar.close()
```
